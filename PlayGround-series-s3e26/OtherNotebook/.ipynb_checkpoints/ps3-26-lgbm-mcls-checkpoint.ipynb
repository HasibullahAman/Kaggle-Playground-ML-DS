# 5. モデルの作成
## Modeling

```python
# trainを学習用データセットと検証用データセットに分割

X = df_train[feat_test]
value = df_train['Status']
```

```python
# 学習用データと検証用データを作成する関数

def make_lgb_data(test_size, random_state, metric, X, value):
    X_train, X_test, t_train, t_test = train_test_split(
        X,
        value,
        test_size=test_size,
        random_state=random_state
    )

    lgb_train = lgb.Dataset(
        X_train,
        t_train
    )
    
    lgb_eval = lgb.Dataset(
        X_test,
        t_test,
        reference=lgb_train
    )
    
    dic_return = {
        'X_train' : X_train,
        'X_test' : X_test,
        't_train' : t_train,
        't_test' : t_test,
        'lgb_train' : lgb_train,
        'lgb_eval' : lgb_eval
    }
    
    return dic_return
```

```python
# optuna を使ってパラメーターチューニングする関数

def tuneParams(test_size, random_state, objective, metric, num_class, X, value):

    opt_params = {
        #'device' : 'gpu',
        'objective' : objective,
        'metric' : metric,
        'num_class' : num_class
    }
    
    dic = make_lgb_data(test_size, random_state, metric, X, value)
    lgb_train = dic['lgb_train']
    lgb_eval = dic['lgb_eval']
    
    opt = lgbo.train(
        opt_params,
        lgb_train,
        valid_sets = lgb_eval,
        #verbose_eval = False,
        num_boost_round = 100#,
        #early_stopping_rounds = 100
    )
    
    return opt
```

```python
# 学習（モデル作成）関数

def make_model(X, value, test_size, random_state, objective, metric, num_class, learning_rate, num_iterations, max_depth, paramObj):
    dic = make_lgb_data(test_size, random_state, metric, X, value)
    lgb_train = dic['lgb_train']
    lgb_eval = dic['lgb_eval']
    X_test = dic['X_test'] # 検証用
    t_test = dic['t_test'] # 　〃

    params = {
        #'device': 'gpu',
        'task': 'train',
        'objective': objective,
        'metric': metric,
        'num_class': num_class,
        'boosting_type': 'gbdt',
        'learning_rate': learning_rate,
        'num_iterations': num_iterations,
        'max_depth': max_depth,
        'feature_pre_filter': paramObj['feature_pre_filter'],
        'lambda_l1': paramObj['lambda_l1'],
        'lambda_l2': paramObj['lambda_l2'],
        'num_leaves': paramObj['num_leaves'],
        'feature_fraction': paramObj['feature_fraction'],
        'bagging_fraction': paramObj['bagging_fraction'],
        'bagging_freq': paramObj['bagging_freq'],
        'min_child_samples': paramObj['min_child_samples'],
        'verbosity': -1
    }

    evaluation_results = {}                       # 学習の経過を保存する
    model = lgb.train(
        params,
        valid_names=['train', 'valid'],           # 学習経過で表示する名称
        valid_sets=[lgb_train, lgb_eval],         # モデル検証のデータセット
        evals_result=evaluation_results,          # 学習の経過を保存
        train_set=lgb_train,
        callbacks=[lgb.early_stopping(1000),
                   lgb.record_evaluation(evaluation_results),
                   lgb.log_evaluation(1000) ]
    )
    
    resultObj = {'evaluation_results' : evaluation_results,
                 'model' : model,
                 'X_test' : X_test, # 検証用
                 't_test' : t_test} #   〃
    return resultObj
```

```python
#----------------------------------------------------
# 基本設定（Base setting）
#----------------------------------------------------
test_size = 0.2
random_state_ary = [0,1,2,3,4,5,6,7,8]
objective = 'multiclass'
metric = 'multi_logloss'# multi_logloss, auc_mu
num_class = 3 # FIX

#----------------------------------------------------
# for optuna
#----------------------------------------------------
optuna_switch = 'off'
opt_count = 1
num_choose = 1 # FIX

if opt_count < num_choose:
    num_choose = opt_count

#----------------------------------------------------
# for lightGBM
#----------------------------------------------------
learning_rate = 0.0005 # 0.0005
num_iterations = 300000
max_depth = -1
```

```python
# optuna

opt_ary = []
if optuna_switch == 'on':
    count_rs = 0
    for random_state in random_state_ary:
        param_ary = []
        for i in range(opt_count):
            count_rs += 1
            print()
            print('=' * 100)
            print(f'Round : {i + 1} / {opt_count} (random_state : {random_state}) ---------->')
            print(f'(Total Round {count_rs} / {len(random_state_ary) * opt_count})')
            print()
            opt = tuneParams(test_size, random_state, objective, metric, num_class, X, value)
            score = opt.best_score['valid_0'][metric]
            dic = {'score' : score, 'params' : opt.params}
            param_ary.append(dic)

        # スコアの高い順にソート
        #param_ary = sorted(param_ary, key=lambda x: x['score'], reverse=True)
        # スコアの低い順にソート
        param_ary = sorted(param_ary, key=lambda x: x['score'], reverse=False)
        
        opt_ary.append({'random_state' : random_state, 'param_ary' : param_ary})
```

```python
# Feature importance
def show_FI(model, random_state, count):
    fig, ax = plt.subplots(figsize=(10, 10))
    lgb.plot_importance(model, ax=ax)
    plt.title('random_state : ' + str(random_state) + ' (Round' + str(count) + ')')
```

```python
# 混同行列
# X軸：予測（Predict）、Y軸：正解（True）

def show_cm(X_test, t_test):
    print()
    print('Confusion Matrix')
    cm = confusion_matrix(t_test, pred_max)
    class_names = ['C', 'CL', 'D']
    sns.heatmap(cm, annot=True, cmap='Blues', xticklabels=class_names, yticklabels=class_names, fmt=',')
    plt.xlabel('Predict')
    plt.ylabel('True')
    plt.show()
```

```python
# 分類結果と精度

def show_report(t_test, pred_max):
    print()
    print('Class Report')
    classReport = classification_report(t_test, pred_max)
    print(classReport)

    score = accuracy_score(t_test, pred_max)
    print(f'accuracy : {score:.5f}')
```

```python
# 検証用データをモデルに投入（モデル別の精度検証）
#（random_stateが異なるとX_testとt_testの組み合わせが異なるのでアンサンブルはできない）

for i in range(len(opt_ary)):
    result_ary = opt_ary[i]['result_ary']
    model = result_ary['model']
    X_test = result_ary['X_test']
    random_state = opt_ary[i]['random_state']

    pred = model.predict(X_test, num_iteration=model.best_iteration)
    # 一番大きい値の位置を取得
    pred_max = np.argmax(pred, axis=1)
    
    t_test = result_ary['t_test']
    
    print()
    print('='*100)
    print(f'random_state : {random_state}')
    show_cm(X_test, t_test)
    show_report(t_test, pred_max)
```
